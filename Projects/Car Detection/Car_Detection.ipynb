```python
#Imports 
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_dir = 'dataset/train'
```

```python
test_dir = 'dataset/test'
```

```python
# Processing the images
train_datagen = ImageDataGenerator(rescale =1./255, validation_split = 0.2) #20% validation
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Load the training data from the local director
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128), # Resize all images 128 x 128
    batch_size = 32,
    class_mode = 'sparse',
    subset = 'training'
)
```

```python
# Load the validation data
validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='sparse',
    subset='validation'
)
```

```python
# Load the test data
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='sparse'
)
```

# CNN Architecture

## Convolution
Using three convolutional layers in a convolutional neural network (CNN) is based on the idea of learning a hierarchical set of features from the image. As data passes through the network, each convolutional layer extracts increasingly complex representations. Here's a breakdown of why three convolutions are used and their purpose:

1. Progressive Feature Learning (Hierarchy)
Each convolutional layer captures progressively more detailed information about the image:

- First Convolutional Layer:
Learns to detect basic features such as edges, lines, and simple textures. These are low-level features that help define the outline and structure of objects in the image.
- Second Convolutional Layer:
Detects more complex features by combining the edges and textures learned in the first layer. This might include recognizing shapes or parts of an object, like the curves of a wheel or parts of a car.
- Third Convolutional Layer:
Identifies high-level features, combining the simpler patterns from earlier layers into more complex representations. At this stage, the model may start recognizing complete objects or large portions of an object (such as entire cars or their major components).

2. Increasing Complexity with More Filters
As you go deeper into the network, the number of filters (kernels) typically increases (32 → 64 → 128). This enables the model to capture a broader range of features at each level of complexity. More filters allow the network to learn more specific patterns, but they also increase the depth of the model, enabling it to recognize more complex shapes and textures.

3. Gradual Spatial Dimensionality Reduction
With each MaxPooling operation following a convolution, the spatial size (height and width) of the image decreases. This is useful because:

- It reduces the amount of data to be processed in the next layer, making computation more efficient.
- It retains only the most important features of the image, removing unnecessary details.
For example:

Input: Original image size of 128x128x3 (RGB).
- After the first convolution and pooling: 64x64x32 (extracts simple features like edges).
- After the second convolution and pooling: 32x32x64 (extracts more complex features).
- After the third convolution and pooling: 16x16x128 (extracts high-level features).

## Overview of the Architecture

1. Sequential Model: tf.keras.models.Sequential
This defines a sequential model, where each layer is stacked linearly one after the other. The model processes input from one end to the other, moving through the layers sequentially.

2. First Convolutional layer: tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3))
- Conv2D(32, (3, 3)): This is a convolutional layer that applies 32 filters of size 3x3 to the input image.
- 32 filters means that the layer will learn 32 different feature maps, detecting different patterns such as edges, corners, or textures.
- input_shape=(128, 128, 3): The input images have a resolution of 128x128 pixels with 3 channels (RGB, meaning color images).
- activation='relu': The ReLU (Rectified Linear Unit) activation function introduces non-linearity, making the model capable of learning complex patterns. ReLU is used to avoid negative values and allow the model to compute faster.

3. First MaxPooling layer: tf.keras.layers.MaxPooling2D(2, 2)
- MaxPooling: This reduces the spatial dimensions (height and width) of the feature maps while retaining important information. Pooling helps the network become invariant to small translations of the input.
- (2, 2): This means a 2x2 window is used for pooling, reducing the width and height of the image by half.

4. Second Convolutional Layer: tf.keras.layers.Conv2D(64, (3, 3), activation='relu')
- Conv2D(64, (3, 3)): This layer applies 64 filters of size 3x3. As the depth of the network increases, the number of filters also increases (32 → 64), allowing the model to capture more complex features from the images.
- activation='relu': Again, ReLU is used to add non-linearity and speed up computation.

5. Second MaxPooling Layer: tf.keras.layers.MaxPooling2D(2, 2)
- Similar to the first pooling layer, this reduces the spatial size of the feature maps, maintaining only the most important information while reducing computational complexity.

6. Third Convolutional Layer: tf.keras.layers.Conv2D(128, (3, 3), activation='relu')
- Conv2D(128, (3, 3)): Here, 128 filters are applied, further increasing the depth of the network. The deeper layers capture even more complex patterns in the image, such as large shapes or combinations of previously detected edges and textures.

7. Third MaxPooling Layer: tf.keras.layers.MaxPooling2D(2, 2)
Reduces the spatial dimensions again, preparing the data for the fully connected layers by flattening the image while retaining learned features.

8. Flatten Layer: tf.keras.layers.Flatten()
- This layer converts the 2D matrix of features into a 1D vector. After all the convolution and pooling operations, the feature maps are flattened into a long vector, which is needed to feed into the fully connected layers (Dense layers).

9. First Dense Layer (Fully Connected Layer): tf.keras.layers.Dense(512, activation='relu')
- Dense(512): This is a fully connected layer with 512 neurons. This layer processes the flattened feature vector from the previous layer and learns to combine the extracted features.
- activation='relu': ReLU is used again to allow for non-linear transformations and faster learning.

10. Output Layer (Classification): tf.keras.layers.Dense(len(train_generator.class_indices), activation='softmax')
- Dense(len(train_generator.class_indices)): This is the output layer, where the number of neurons corresponds to the number of classes in your classification problem (e.g., different car types). The len(train_generator.class_indices) dynamically sets the number of neurons based on the number of classes in the dataset.
- activation='softmax': The softmax function is used for multi-class classification. It converts the raw output into a probability distribution over the classes. The class with the highest probability is the model’s prediction.


```python
# Define your CNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(len(train_generator.class_indices), activation='softmax')  # Output layer
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10
)
```

```python
# Evaluate on the test set
model.evaluate(test_generator)
```

```python
# Function to predict a single image
def predict_car_image(image_path):
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(128, 128))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Create batch dimension

    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions)

    # Get class label name from the class_indices
    label_name = {v: k for k, v in train_generator.class_indices.items()}[predicted_class]
    
    # Plot the image and predicted class
    plt.imshow(img)
    plt.title(f"Predicted: {label_name}")
    plt.axis('off')
    plt.show()
```

```python
predict_car_image('dataset/validation/bug.jpg')
```
